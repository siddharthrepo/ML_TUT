```python
sns.countplot(x = 'Gender' , data=df)
```

```python
ax = sns.countplot(x="Gender" , data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'] , as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x = "Gender" , y="Amount" , data = sales_gen)
```

```python
sns.countplot(data=df , x = "Age Group" )
```

```python
sns.countplot(data=df , x = "Age Group" , hue="Gender")
```

```python
ax= sns.countplot(data=df , x = "Age Group" , hue="Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'] , as_index=False)['Amount'].sum().sort_values(by="Amount", ascending=False)
sns.barplot(x='Age Group' , y='Amount' , data = sales_age)
```

```python
sales_state = df.groupby(['State'] , as_index=False)["Orders"].sum().sort_values(by="Orders" , ascending=False)
sns.set(rc={'figure.figsize':(15,10)})  # if this is not given , space between graphs will not be clear
sns.barplot(data = sales_state , x="State" , y="Orders")
```

```python
ax= sns.countplot(data=df , x = "Shaadi")
sns.set(rc={'figure.figsize':(7,5)})
for bars in ax.containers:
    ax.bar_label(bars)

```

```python
sns.set(rc={'figure.figsize':(20,8)})
ax = sns.countplot(data=df , x="Occupation")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
purchasing_power = df.groupby(['Occupation'] , as_index=False)['Amount'].sum().sort_values(by="Amount" ,ascending=False)
sns.set(rc={'figure.figsize' : (20,5)})
sns.barplot(data= purchasing_power , x= "Occupation")
```

```python
sns.set(rc={"figure.figsize":(25,10)})
ax = sns.countplot(data = df , x='Product_Category' )

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
orders = df.groupby(['Product_Category'] , as_index=False)['Amount'].sum().sort_values(by="Amount" , ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,10)})
sns.barplot(data = orders , x='Product_Category' , y="Amount")

```
